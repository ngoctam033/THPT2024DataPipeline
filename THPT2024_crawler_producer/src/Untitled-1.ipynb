```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu mẫu
data = {
    'Mã Học Sinh': ['37000001', '37000002', '37000003', '37000004', '37000005'],
    'Tên': ['Nguyen Van A', 'Tran Thi B', 'Le Van C', 'Pham Thi D', 'Hoang Van E'],
    'Toán': [9.0, 7.5, 8.5, 9.5, 8.0],
    'Ngữ văn': [8.0, 6.5, 8.0, 9.0, 7.5],
    'Ngoại ngữ': [7.5, 8.5, 7.0, 9.5, 8.0],
    'Vật lý': [8.0, 7.0, 9.0, 9.5, 7.5],
    'Hóa học': [9.5, 8.0, 7.5, 9.0, 8.5],
    'Sinh học': [7.0, 8.5, 8.0, 9.5, 7.5]
}

# Tạo DataFrame từ dữ liệu
df = pd.DataFrame(data)

# Tính tổng điểm cho mỗi sinh viên
df['Tổng điểm'] = df[['Toán', 'Ngữ văn', 'Ngoại ngữ', 'Vật lý', 'Hóa học', 'Sinh học']].sum(axis=1)

# Sắp xếp sinh viên theo tổng điểm từ cao đến thấp
df = df.sort_values(by='Tổng điểm', ascending=False)

# Lấy top 5 sinh viên có điểm số cao nhất
top_students = df.head(5)

# Vẽ biểu đồ cột thể hiện xếp hạng sinh viên
plt.figure(figsize=(10, 6))
plt.bar(top_students['Tên'], top_students['Tổng điểm'], color='skyblue')
plt.xlabel('Tên Sinh Viên')
plt.ylabel('Tổng Điểm')
plt.title('Top 5 Sinh Viên Có Tổng Điểm Cao Nhất')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from selenium.webdriver.common.by import By
def find_element_by_css(bot, css_selector):
    """Tìm và trả về phần tử bằng CSS selector"""
    try:
        element = bot.driver.find_element(By.CSS_SELECTOR, css_selector)
        return element
    except Exception as e:
        print(f"Không thể tìm thấy phần tử: {e}")
        return None
```
